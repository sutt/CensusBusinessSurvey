```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
books = pd.read_csv('../data/goodreads-10k/books.csv')
```

```python
# convert float to int with na values
books['original_publication_year'] = books['original_publication_year'].fillna(0)
books['original_publication_year'] = books['original_publication_year'].astype(int)
```

```python
books['new_col'] = books['original_publication_year'].map(lambda x: int(x) if pd.isna(x) else x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30, edgecolor='black')

# Set axes labels
plt.xlabel('Value')
plt.ylabel('Frequency')

# Get the current axes
ax = plt.gca()

# Set axes label background color and alpha (transparency)
ax.xaxis.set_bbox(dict(facecolor='white', alpha=0.5, edgecolor='none'))
ax.yaxis.set_bbox(dict(facecolor='white', alpha=0.5, edgecolor='none'))

# Display the plot
plt.show()
```
