```python
data = {
        'a': [1,2,1],
        'b': ["a", "b", "c"],
        'c': [22.3, 1, -10],
}

df = pd.DataFrame(data)

display(df)
display(df)
```

```python
df2 = df.copy()
```

```python
df2['c'] = 1.0
```

```python
bigdf = pd.concat([df, df2], axis=0)
bigdf
```

```python
bigdf.index = pd.RangeIndex(start = 1, stop=18,step=3)
```

```python
bigdf.reset_index(inplace=True, drop=True)
bigdf
```

```python
bigdf['id'] = bigdf.apply(lambda row: row['a'] + 10, axis=1)
bigdf
```

```python
for i in range(len(bigdf)):
    bigdf['id'][i] = bigdf.iloc[ i, 0] + 10
bigdf
```

```python
bigdf['index_at_one'] = bigdf.index
bigdf['index_at_one'] = bigdf['index_at_one'].map(lambda x: x+1)
bigdf.index = bigdf['index_at_one']
```
