```python
import pandas as pd
goodreads = pd.read_csv('../data/goodreads-10k/books.csv')
```

Decide how to handle empty values in the following columns:

- ISBN
- ISBN13
- Language

```python
# Not handling empty values unless absolutely needed to be addressed
goodreads.info()

# Filtering fields

goodreads = goodreads[['title','original_title','authors','original_publication_year','isbn13']]
```

- Add a column to show where original\_title and title differ.

```python
goodreads['title_mismatch'] = goodreads['original_title'] != goodreads['title']
goodreads.head()
```

- Format ISBN 13 to match the specification with dashes. Consult [the ISBN information page](https://isbn-information.com/the-13-digit-isbn.html) for guidance.

3-1-6-2-1

```python
# Drop NA
cleaned_isbn13_data = goodreads.dropna(subset=['isbn13'])

cleaned_isbn13_data.drop(['title','original_title','authors','original_publication_year','title_mismatch'],axis=1,inplace=True)

# Cleaning fields
cleaned_isbn13_data['isbn13str'] = cleaned_isbn13_data['isbn13'].astype(str)

# Convert to string
cleaned_isbn13_data['isbn13strlen'] = cleaned_isbn13_data['isbn13str'].apply(len)

# Check data
cleaned_isbn13_data['isbn13strlen'].value_counts()
```

```python
# What are the titles less than 15 characters for ISBN 13?
cleaned_isbn13_data[cleaned_isbn13_data['isbn13strlen'] < 15]

# 3 - 1 - 6 - 2 - 1

cleaned_isbn13_data['isbn13str1'] = cleaned_isbn13_data['isbn13str'].str.slice(0,3)
cleaned_isbn13_data['isbn13str2'] =cleaned_isbn13_data['isbn13str'].str.slice(3,4)
cleaned_isbn13_data['isbn13str3'] =cleaned_isbn13_data['isbn13str'].str.slice(4,10)
cleaned_isbn13_data['isbn13str4'] =cleaned_isbn13_data['isbn13str'].str.slice(10,12)
cleaned_isbn13_data['isbn13str5'] =cleaned_isbn13_data['isbn13str'].str.slice(12,13)
cleaned_isbn13_data['isbn13formatted'] = cleaned_isbn13_data['isbn13str1']+"-"+cleaned_isbn13_data['isbn13str2']+"-"+cleaned_isbn13_data['isbn13str3']+"-"+cleaned_isbn13_data['isbn13str4']+"-"+cleaned_isbn13_data['isbn13str5']
cleaned_isbn13_data.drop(['isbn13strlen','isbn13str','isbn13str1','isbn13str2','isbn13str3','isbn13str4','isbn13str5'],axis=1,inplace=True)
goodreads = goodreads.merge(cleaned_isbn13_data,how='inner',on=['isbn13'])
```

```python
goodreads.head()
```

- Break up the authors into separate columns.

```python
goodreads_authorsplit = goodreads.authors.str.split(',',expand=True).add_prefix('author')
goodreads.drop('authors',axis=1,inplace=True)
goodreads = goodreads.join(goodreads_authorsplit)
```

- Store each publication by its publication year by century in its own worksheet. So 1800-1899, 1900-1999, 2000-2099, etc.

```python
goodreads.dropna(subset=['original_publication_year'],inplace=True)
goodreads['original_publication_year'] = goodreads['original_publication_year'].astype(str).str.replace('\.0','').astype(int)
```

```python
starting_year = goodreads['original_publication_year'].min()-50
ending_year = goodreads['original_publication_year'].max()
step = 100
with pd.ExcelWriter('books_by_century.xlsx') as xlsxWriter:
    for time_period in range(starting_year,ending_year,step):
        filtered_set = goodreads[(goodreads.original_publication_year > time_period) & (goodreads.original_publication_year <= time_period + 100)]
        filtered_set = filtered_set.dropna(how='all',axis=1)
        if len(filtered_set) > 0:
            filtered_set.to_excel(xlsxWriter,sheet_name=f"{time_period}-{time_period+100}")
xlsxWriter.save()
```
